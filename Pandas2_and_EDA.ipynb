## GroupBY

```python
import pandas as pd
df = pd.read_csv('/content/sample_data/weather_by_cities.csv')
print(df)
```

```python
gb=df.groupby('city')
gb
```

```python
for city,data in gb:
  print("city",city)
  print()
  print(data)
  print()
```

```python
gb.get_group('mumbai')#will return dataframe
```

```python
gb.max()#gives all max values
```

```python
gb.mean()
```

```python
gb.describe() #descriptive statistics
```

```python
gb.size() #total no.of values in each group
```

```python
gb.count()
```

##concatination

```python
ind_weather=pd.DataFrame({
    "city":['mumbai','hyd','delhi'],
    "temp":[32,45,50],
    "humidity":[50,70,80]
    })
ind_weather
```

```python
US_weather=pd.DataFrame({
    "city":['usa','japan','canada'],
    "temp":[23,34,21],
    "humidity":[50,60,100]
    })
US_weather
```

```python
df=pd.concat([ind_weather,US_weather],ignore_index='True')
df
```

```python
df=pd.concat([ind_weather,US_weather],keys=['india','us'])
df
```

```python
df.loc['us']
```

```python
type(df.loc['us'])
```

```python
df=pd.concat([ind_weather,US_weather],axis=1)
df
```

```python
df[['city']]
```

```python
s=pd.Series(['humid','dry','rain'],name='Event')
s
```

```python
df=pd.concat([ind_weather,s],axis=1)
df
```
