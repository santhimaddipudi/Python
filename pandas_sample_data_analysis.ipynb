```python
Sales_data=Sales_data.dropna(how='all')
Sales_data=Sales_data[Sales_data['Order Date'].str[0:2]!='Or']
Sales_data['Month']=Sales_data['Order Date'].str[0:2]
#df1['Month']=df1['Month'].astype['int32']
Sales_data.head()
```

```python
Sales_data['Quantity Ordered']=Sales_data['Quantity Ordered'].astype(int)
Sales_data['Price Each']=Sales_data['Price Each'].astype(float)
```

```python
Sales_data['Sales']=Sales_data['Quantity Ordered']*Sales_data['Price Each']
Sales_data.head()
```

```python
#What was the best month for sales
Sales_data.groupby('Month').sum()
```

```python
import matplotlib.pyplot as plt
months=range(1,13) #x-axis
results=Sales_data.groupby('Month').sum()
plt.bar(months,results['Sales'])
plt.xticks(months)
labels,location=plt.yticks()
plt.yticks(labels,(labels/1000000).astype(int)) #scaling in million USD
plt.ylabel("Sales in USD")
plt.xlabel('Month Number')
plt.show()
```
